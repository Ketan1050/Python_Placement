```python
import csv 
with open("data.csv", "w") as f:
    writer = csv.writer(f) # csv.writer is a function which returns a writer object which can be used to write to the file.
    writer.writerow(["name", "age", "city"]) # writerow is a method which writes a row to the file. It takes a list as an argument. Each element of the list is written to the file as a separate column.
    writer.writerow(["ketan", 20, "mumbai"])
    writer.writerow(["john", 25, "new york"]) 
with open("data.csv", "r") as f:
    reader = csv.reader(f) # csv.reader is a function which returns a reader object which can be used to read the file. 
    for row in reader:
        print(row) 
    
    
```
