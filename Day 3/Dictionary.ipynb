```python
translate ={} 
# populate the dictionary 

translate["danke"] = "Thank you"  
translate["coffee"] = "kaffee" 
translate["milk"] = "milch" 
translate["bye"] = "tschuss" 
translate["dog"] = "hund" 
translate["cat"] = "katzee" 

def eng2german(word): 
    if word not in translate: 
        return "word not found " 
    return translate[word] 

print(eng2german("bye")) 
```

```python
emoji = { 
        "happy" : "😊", 
        "love" : "❤️", 
        } 

word = input("Mood") 
sentence = "I am " 
print(sentence + emoji[word])   

```

```python
def replace_mood(sentence): 
    words = sentence.split() 
    result = "" 
    for i in words: 
        if i in emojis: 
            result += emojis[i] 
        else: 
            result += i 
    return result 

emojis = { 
        "happy" : "😊", 
        "love" : "❤️", 
        } 

print("I am ") 
sentence = input("Enter a sentence: ") 
print(replace_mood(sentence)) 
```
