```python
def volume(l = 1, b = 2, h = 1): 
    return l * b * h 

print(volume()) #function calling 
print(volume(1,2,3)) 
print(volume(h=7, b=6, l=3)) 
```

```python
import math 
def hypotenuse(a,b): 
    
    return math.sqrt(a * a + b * b) 
print(hypotenuse(3,4)) 
 
```

##Variable Length Argument

```python
def f(*x): 
    return x #here x is tuple 

# calling function 
print((1,2,3))   
```

```python
def average(*n) :  
    return sum(n)/len(n) #n is tuple 

print(average(1,2,3,4,5))  
```

```python
def avarage(*x): 
    s = 0 
    for i in x: 
        s = i + 1 
    return s/len(x) 

print(avarage(1,2))  
```
