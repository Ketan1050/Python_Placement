List Comprehension


```python
#[body_of_for for_loop] 
#[body_of_if for_loop if condition]  
```

```python
[i**2 for i in range(1,10,2)]  # 1 + 2 
```

```python
print(round(1.262373,0)) 
print(round(1.262373,1))  
print(round(1.262373,2)) 
print(round(1.262373,3)) 
```

```python
import math as m 

[round(m.pi, i) for i in range(1,3)] #m.pi is in for loop and i is in body of for loop. 
```

```python
[ord(i) for i in "ketan"] # ord gives the unicode value of the character. 
```

```python
chr(ord("k")-32) # chr gives the character of the unicode value. 32 is the difference between small and capital letters in unicode.
```
